````python
# Replace with your dataset (e.g., Yahoo Finance CSV)
df = pd.read_csv("stock_yfinance_data.csv")

# Inspect dataset
print(df.head())
print(df.info())

# Convert date column if exists
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
````

````python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['Close']].values)

LOOKBACK = 60  # 60-day window
X, y = [], []
for i in range(LOOKBACK, len(scaled)):
    X.append(scaled[i-LOOKBACK:i, 0])
    y.append(scaled[i, 0])
X, y = np.array(X), np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Train-test split
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
````

````python
model = Sequential([
    LSTM(128, return_sequences=True, input_shape=(X.shape[1], 1)),
    Dropout(0.2),
    LSTM(64),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")
model.summary()
````

````python
history = model.fit(
    X_train, y_train,
    epochs=20,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)
````

````python
y_pred = model.predict(X_test)

# Rescale back
y_pred_rescaled = scaler.inverse_transform(y_pred)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1,1))

# Metrics
rmse = np.sqrt(mean_squared_error(y_test_rescaled, y_pred_rescaled))
mape = mean_absolute_percentage_error(y_test_rescaled, y_pred_rescaled) * 100

print("RMSE:", rmse)
print("MAPE:", mape)
````

````python
plt.figure(figsize=(12,6))
plt.plot(y_test_rescaled, label="Actual", color="black")
plt.plot(y_pred_rescaled, label="Predicted", color="red")
plt.legend()
plt.title("Stock Price Prediction")
plt.show()
````
